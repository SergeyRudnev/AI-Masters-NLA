- (3 pts) Докажите, что $\| A \|_F \le \sqrt{\mathrm{rank}(A)} \| A \|_2$.

$$\|A\|_F = \Big(\sum_{j=1}^m \sum_{i=1}^n | a_{ij}|^2\Big)^{1/2} = \Big(\sum_{j=1}^m (\bar a_{j}^T, a_j) \Big)^{1/2} = \sqrt{\mathrm{tr}(A^*A)},$$ так как элементы матрицы $A^*A$ - скалярные произведения столбцов матрицы $A$

Так как при замене базиса характеристический многочлен матрицы линейного оператора не меняется и так как коэффициент при $\lambda ^{n−1}$ характеристического многочлена равен произведению $(−1)^{n−1}$ на след, то след - инвариант при замене базиса. 

Поэтому $\mathrm{tr}(A^*A) = \sum_{i=1}^m \sigma_i^2,$ где $A^*A$ - матрица $m \times m$, $\sigma_i^2$ - собственные значения $A^*A$

Также в доказательстве сингулярного разложения было показано, что $r:=\mathrm{rank}(A)$ равен количеству ненулевых $\sigma_i$. $\mathrm{rank}(A) \le m$ 

Отсюда следует, что $$\|A\|_F = \sqrt{\sum_{i=1}^m \sigma_i^2} = \sqrt{\sum_{i=1}^r \sigma_i^2} = \sqrt{\sum_{i=1}^r \lambda_i(A^*A)}\le  \sqrt{r\lambda_{max}(A^*A)} = \sqrt{r} ||A||_2$$

Так как $\|A\|_F = \sqrt{\sum_{i=1}^{\mathrm{rank}(A)} \lambda_i(A^*A)}$, то для равенства будет достаточно того, чтобы все ненулевые сингулярные числа $\sigma_i$ совпадали

Для этого в качестве матрицы $A$ можно взять единичную матрицу размера $k\times k$ и дополнить её нулями до матрицы размера $m \times n$

````python
# INPUT : diag_broadcast - list of diagonals value to broadcast, length equal to 3 or 5; n - integer, band matrix shape.
# OUTPUT : L - 2D np.ndarray, L.shape[0] depends on bandwidth, L.shape[1] = n-1, do not store main diagonal, where all ones;                  add zeros to the right side of rows to handle with changing length of diagonals.
#          U - 2D np.ndarray, U.shape[0] = n, U.shape[1] depends on bandwidth;
#              add zeros to the bottom of columns to handle with changing length of diagonals.
def band_lu(diag_broadcast, n):
    if(len(diag_broadcast) == 3):
        u = np.zeros(n)
        l = np.zeros(n-1)
        lamda = np.zeros(n) + diag_broadcast[2]
        lamda[-1] = np.NaN
        u[0] = diag_broadcast[1]
        for i in range(n-1):
            l[i] = diag_broadcast[0] / u[i]
            u[i+1] = diag_broadcast[1] - l[i] * diag_broadcast[2]
        return np.array([l]), np.array([u, lamda]).T
    if(len(diag_broadcast) == 5):
        u = np.zeros(n)
        lamda = np.zeros(n)
        lamda[-1] = np.NaN
        alpha = np.zeros(n) + diag_broadcast[4]
        alpha[-2:] = np.NaN 
        l = np.zeros(n-1)
        beta = np.zeros(n-1)
        beta[-1] = np.NaN
        u[0] = diag_broadcast[2]
        lamda[0] = diag_broadcast[3]
        l[0] = diag_broadcast[1] / u[0]
        u[1] = diag_broadcast[2] - l[0] * lamda[0]
        lamda[1] = diag_broadcast[3] - l[0] * diag_broadcast[4]
        for i in range(n-3):
            beta[i] = diag_broadcast[0] / u[i]
            l[i+1] = (diag_broadcast[1] - beta[i] * lamda[i]) / u[i+1]
            u[i+2] = diag_broadcast[2] - beta[i] * diag_broadcast[4] - l[i+1] * lamda[i+1]
            lamda[i+2] = diag_broadcast[3] - l[i+1] * diag_broadcast[4]
        beta[n-3] = diag_broadcast[0] / u[n-3]
        l[n-2] = (diag_broadcast[1] - beta[n-3] * lamda[n-3]) / u[n-2]
        u[n-1] = diag_broadcast[2] - beta[n-3] * diag_broadcast[4] - l[n-2] * lamda[n-2]
        return np.array([l, beta]), np.array([u, lamda, alpha]).T
````

````python
def LU(Matrix):
    A = Matrix.copy()
    n = A.shape[0]
    L = np.zeros((n, n))
    U = np.zeros((n, n))
    for k in range(n):
        L[k, k] = 1
        for i in range(k+1, n):
            L[i, k] = A[i, k] / A[k, k]
            for j in range(k+1, n):
                A[i, j] = A[i, j] - L[i, k] * A[k, j]
        for j in range(k, n):
            U[k, j] = A[k, j]
    return L, U
````

Воспользуемя алгоритмом с лекции, который каждый раз выборает максимальный по модулю элемент в строке и меняет столбцы местами так, чтобы этот элемент оказался на диагонали. Замена столбцов эквивалентна умножению справа на метрицу перестановки.
Получим:
$L^{-1}AP = U$ Отсюда 
$ A = LUP^{-1} = LUP^T = LUQ$, так как матрица перестановки ортогональна

````python
def swap_columns(Matrix, i_,j_):
    for m in range(Matrix.shape[0]):
        buf = Matrix[m, i_]
        Matrix[m, i_] = Matrix[m, j_]
        Matrix[m, j_] = buf
````

````python
def LUQ(Matrix):
    a = Matrix.copy()
    n = a.shape[0]
    L = np.zeros((n, n))
    U = np.zeros((n, n))
    P = np.identity((n))

    for k in range(n):
        pivot = a[k, k]
        pivot_ind = k
        for l in range(k+1, n):
            if np.abs(a[k, l]) > np.abs(pivot):
                pivot = a[k, l]
                pivot_ind = l
        if pivot_ind != k:
            swap_columns(P, k, pivot_ind)
            swap_columns(a, k, pivot_ind)
        L[k, k] = 1
        for i in range(k+1, n):
            L[i, k] = a[i, k] / a[k, k]
            for j in range(k+1, n):
                a[i, j] = a[i, j] - L[i, k] * a[k, j]
    for k in range(n):
        for j in range(k, n):
            U[k, j] = a[k, j]
    return L, U, P.T
````

````python
Matrix = np.array([[-5, 4, 0, 0],[3,  -5, 4, 0], [0, 3, -5, 4], [0, 0, 3, -5]], dtype = float)
L, U, Q = LUQ(Matrix)
np.linalg.norm(Matrix - L @ U @ Q, 'fro')
````

````python
eps = 1e-16
Matrix = np.array([[eps, 1, 0],[1, 1, 1], [0, 1, 1]], dtype = float)
L, U, Q = LUQ(Matrix)
np.linalg.norm(Matrix - L @ U @ Q, 'fro')
````

Объясним, почему алгоритм вычисления $LU$ разложения без выбора ведущего элемента плохо аппроксимирует факторы  $𝐿$  и  $𝑈$ для  $|𝜀|≪1.$

Заметим, что при вычислении $U_{2,2}$ компьютер получит, что $U_{2,2} = 1 - 𝜀^{-1} = -𝜀^{-1}$.
Также при вычислении $L_{3,2}$ получим $L_{3,2} = 1 / U_{2,2} = -𝜀$. И, следовательно, $U_{3,3} = 1 +  𝜀 = 𝜀.$ Из-за этих округлений и возникает ошибка. А если бы мы не делили на 𝜀 на первом шаге, то таких ошибок бы не возникло, о чём свидетельствует метод $LUQ$

Из этих двух равенств и получаем нужное

Пользовались фактом, что $$\det\Big(\begin{bmatrix} B & 0 \\ 0 & C \end{bmatrix}\Big) = \det(B) \det(C)$$

Первый факт можно свести к доказательству второго: $\det A$ является полилинейной
кососимметрической функцией первых $n$ столбцов (матрица $B$ - $n\times n$). То есть $\det A = F(B).$ Тогда в силу определения определителя $\det A = F(B) = F(I)\det(B),$ где $F(I) = \det\Big(\begin{bmatrix} I & 0 \\ 0 & C \end{bmatrix}\Big)$

Также $F(I)$ является полилинейной
и кососимметрической функцией своих последних $m$(матрица $С$ - $m\times m$) строк. Поэтому $F(I) = G(C) = G(I)\det C,$ где 
$G(I) = \det\Big(\begin{bmatrix} I & 0 \\ 0 & I \end{bmatrix}\Big)$

Доказательство того, что определитель верхнетреугольной матрицы равен произведению диагональных элементов, проводится по индукции. Просто раскрываем определитель матрицы размера $n\times n$ по первой строке, получаем произведение первого элемента с диагонали на определитель матрицы размера $(n-1) \times (n-1),$ а он уже по предположению индукции равен птроизведению диагональных элементов. ч.т.д.

Прототип приведён ниже и не должен изменяться. Обратите внимание на эффективность реализации операции умножения матриц Хаусхолдера для получения матрицы ```Q```.

````python
# INPUT : rectangular matrix A
# OUTPUT: matrices Q - orthogonal and R - upper triangular such that A=QR
# A - m*n, Q - m*min(m,n), R - min(m,n)*n
# Q = Q_1^{*}...Q_t^{*}, t = min(m-1,n)
import numpy as np
def scal_prod(a,b):
    res = 0
    for i in range(len(a)):
        res += a[i] * (b[i].conjugate())
    return res.real
def normalize(a):
    return a / np.sqrt(scal_prod(a,a))
def v(x):
    e = np.zeros(len(x))
    e[0] = np.sqrt(scal_prod(x,x))
    return normalize((x + e) / np.sqrt(2*(scal_prod(x,x) - x[0]*np.sqrt(scal_prod(x,x)))))
def Hx(x, v):
    return x - 2*v*scal_prod(v.conjugate(),x)
def xH(x, v):
    return x - 2*v.conjugate()*scal_prod(x,v)
def fill_Q1(v, A):
    Q = np.eye((A.shape[0]))
    for i in range(A.shape[0]):
        for j in range(A.shape[0]):
            Q[i,j] -= 2*v[i]*(v[j].conjugate()) 
    return Q
def householder_qr(A): # 7 pts
    for i in range(np.min([A.shape[0] - 1, A.shape[1]])):
        u = v(A[i:,i])  # вектор, определяющий матрицу H на текущем шаге 
        if i == 0:
            Q = fill_Q1(u, A)
        for j in range(i, A.shape[1]):
            A[i:,j] = Hx(A[i:,j], u)
        if i > 0:
            for l in range(A.shape[0]): # восстанавливаем Q, для этого умножаем справа каждую строку текущей матрицы на следующую матрицу H
                Q[l,i:] = xH(Q[l,i:], u)
    Q = Q[:,:np.min([A.shape[0], A.shape[1]])]
    R = A[:np.min([A.shape[0], A.shape[1]]),:]
    return Q, R
````

### Архитектура нейронной сети

Для простоты и демонстрации принципа сжатия слоёв нейронных сетей рассмотрим архитектуру, состоящую только из полносвязных слоёв (то есть линейных преобразований) и стандартных ReLU-нелинейностей. Опять же для наглядности выбрали размерность 1000 для полносвязных слоёв. 

````python
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(3 * 32 * 32, 1000)
        self.fc2 = nn.Linear(1000, 1000)
        self.fc3 = nn.Linear(1000, 1000)
        self.fc4 = nn.Linear(1000, 1000)
        self.fc5 = nn.Linear(1000, 1000)
        self.fc6 = nn.Linear(1000, 10)
        self.ReLU = nn.ReLU()

    def forward(self, x): # суперпозиция функций от входного вектора
        x = self.fc1(x.view(-1, 3 * 32*32))
        x = self.ReLU(x)

        y = self.fc2(x)
        y = self.ReLU(y)
        y = self.fc3(y)

        x1 = self.ReLU(y + x)

        
        y1 = self.fc4(x1)
        y1 = self.ReLU(y1)
        y1 = self.fc5(y1)

        x2 = self.ReLU(x1 + y1)

        z = self.fc6(x2)
        return F.log_softmax(z, dim=1)
````

#### Функции для обучения и тестирования результата после каждого прохождения всех данных 

````python
def train(model, train_loader, optimizer, epoch):
    model.train()
    for batch_idx, (data, target) in enumerate(train_loader):
        optimizer.zero_grad()
        output = model(data)
        loss = F.nll_loss(output, target)
        loss.backward()  # подсчёт градиентов
        optimizer.step() # обновление параметров
        if batch_idx % log_interval == 0:
            print('Train Epoch: {} [{}/{} ({:.0f}%)]\tLoss: {:.6f}'.format(
                epoch, batch_idx * len(data), len(train_loader.dataset),
                100. * batch_idx / len(train_loader), loss.item()))
````

````python
def test(model, test_loader):
    model.eval()
    test_loss = 0
    correct = 0
    with torch.no_grad():
        for data, target in test_loader:
#             data, target = data.to(device), target.to(device)
            output = model(data)
            test_loss += F.nll_loss(output, target, reduction='sum').item() # sum up batch loss
            pred = output.argmax(dim=1, keepdim=True) # get the index of the max log-probability
            correct += pred.eq(target.view_as(pred)).sum().item()
    test_loss /= len(test_loader.dataset)

    print('\nTest set: Average loss: {:.4f}, Accuracy: {}/{} ({:.0f}%)\n'.format(
        test_loss, correct, len(test_loader.dataset),
        100. * correct / len(test_loader.dataset)))
````

- (12 pts) Создайте модель (аналогично классу ```Net```), которая на вход будет принимать экземпляр класса ```Net```, ранг $r > 0$ и список индексов слоёв, веса которых будут сжиматься. Эта модель должна сжимать некоторые полносвязные слои (индексы которых переданы на вход) с помощью SVD разложения, из которого берутся только первые $r$ компонент. Также функция ```forward``` должна быть эффективно реализована с использованием сжатого представления полносвязных слоёв, а в остальном полностью повторять эту же функцию для несжатой обученной сети.


````python
def CompressedModel(r, layers_ind_to_compress): # layers_ind_to_compress - номера линейных слоёв, которые нужно сжать
    class SVDNet(nn.Module):
        def __init__(self):
            super(SVDNet,  self).__init__()
            self.r = r
            self.layers_ind_to_compress = layers_ind_to_compress
            self.fc1 = nn.Linear(3 * 32 * 32, 1000)
            self.fc1_1 = nn.Linear(3 * 32 * 32, self.r)
            self.fc1_2 = nn.Linear(self.r, self.r)
            self.fc1_3 = nn.Linear(self.r, 1000)
            self.fc2 = nn.Linear(1000, 1000)
            self.fc2_1 = nn.Linear(1000, self.r)
            self.fc2_2 = nn.Linear(self.r, self.r)
            self.fc2_3 = nn.Linear(self.r, 1000)
            self.fc3 = nn.Linear(1000, 1000)
            self.fc3_1 = nn.Linear(1000, self.r)
            self.fc3_2 = nn.Linear(self.r, self.r)
            self.fc3_3 = nn.Linear(self.r, 1000)
            self.fc4 = nn.Linear(1000, 1000)
            self.fc4_1 = nn.Linear(1000, self.r)
            self.fc4_2 = nn.Linear(self.r, self.r)
            self.fc4_3 = nn.Linear(self.r, 1000)
            self.fc5 = nn.Linear(1000, 1000)
            self.fc5_1 = nn.Linear(1000, self.r)
            self.fc5_2 = nn.Linear(self.r, self.r)
            self.fc5_3 = nn.Linear(self.r, 1000)
            self.fc6 = nn.Linear(1000, 10)
            self.fc6_1 = nn.Linear(1000, self.r)
            self.fc6_2 = nn.Linear(self.r, self.r)
            self.fc6_3 = nn.Linear(self.r, 10)
            self.ReLU = nn.ReLU()

        def forward(self, x): # суперпозиция функций от входного вектора
            if 1 in self.layers_ind_to_compress: 
                x = self.fc1_1(x.view(-1, 3 * 32*32))
                x = self.fc1_2(x)
                x = self.fc1_3(x)
            else :
                x = self.fc1(x.view(-1, 3 * 32*32))

            x = self.ReLU(x)
            if 2 in self.layers_ind_to_compress:
                y = self.fc2_1(x)
                y = self.fc2_2(y)
                y = self.fc2_3(y)
            else :
                y = self.fc2(x)

            y = self.ReLU(y)
            if 3 in self.layers_ind_to_compress:
                y = self.fc3_1(y)
                y = self.fc3_2(y)
                y = self.fc3_3(y)
            else:
                y = self.fc3(y)

            x1 = self.ReLU(y + x)

            if 4 in self.layers_ind_to_compress:
                y1 = self.fc4_1(x1)
                y1 = self.fc4_2(y1)
                y1 = self.fc4_3(y1)
            else:  
                y1 = self.fc4(x1)

            y1 = self.ReLU(y1)
            if 5 in self.layers_ind_to_compress:
                y1 = self.fc5_1(y1)
                y1 = self.fc5_2(y1)
                y1 = self.fc5_3(y1)
            else:
                y1 = self.fc5(y1)

            x2 = self.ReLU(x1 + y1)
            if 6 in self.layers_ind_to_compress:
                z = self.fc6_1(x2)
                z = self.fc6_2(z)
                z = self.fc6_3(z)
            else:
                z = self.fc6(x2)
            return F.log_softmax(z, dim=1)
    return SVDNet()
````
